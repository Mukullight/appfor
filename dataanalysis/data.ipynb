```python
import kagglehub
import pandas as pd
import os

# Download latest version of the dataset
path = kagglehub.dataset_download("mohdshahnawazaadil/restaurant-dataset")

# Print the path to the dataset directory
print("Path to dataset files:", path)

# List files in the dataset directory to find the CSV file
for file in os.listdir(path):
    print(file)
    if file.endswith('.csv'):
        csv_path = os.path.join(path, file)
        print(f"Found CSV file: {csv_path}")
        # Read the CSV file
        df = pd.read_csv(csv_path)
        print(df.head())  # Display the first few rows of the DataFrame
```

```python
import plotly.express as px

# Select relevant columns and drop rows with missing values
df_plot = df[['Latitude', 'Longitude', 'Country Code', 'Restaurant Name']].dropna()

# Create an interactive scatter map
fig = px.scatter_geo(df_plot,
                     lat='Latitude',
                     lon='Longitude',
                     color='Country Code',  # Color points by Country Code
                     hover_name='Restaurant Name',  # Show restaurant name on hover
                     hover_data={'Country Code': True, 'Latitude': True, 'Longitude': True},
                     projection='natural earth',  # Use natural earth projection for the map
                     title='Restaurant Locations by Country Code',
                     color_discrete_sequence=px.colors.qualitative.Bold)  # Distinct colors

# Customize the layout with size adjustments
fig.update_layout(
    geo=dict(showframe=False, showcoastlines=True, projection_type='natural earth'),
    legend_title_text='Country Code',
    title_x=0.8,  # Title position as provided
    width=1000,  # Set plot width in pixels
    height=500  # Set plot height in pixels
)

# Display the plot in the Jupyter cell
fig.show()
```
